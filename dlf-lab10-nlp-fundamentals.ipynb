### Key Concepts Covered:

#### Word Embeddings:
- **Word2Vec**: A widely-used model that converts words into vectors, capturing semantic relationships between words. This lab demonstrates how to load pre-trained Word2Vec models and use them to find similar words and analyze word relationships.
- **FastText**: An extension of Word2Vec that represents words as combinations of character n-grams, allowing for better handling of out-of-vocabulary words and morphologically rich languages. The lab includes loading and using pre-trained FastText models.
- **GloVe** (Global Vectors for Word Representation): Another popular word embedding technique that uses co-occurrence statistics to generate word vectors. This lab shows how to load and utilize GloVe embeddings for word similarity tasks.

#### Text Similarity and Distance:
- **Word Mover's Distance (WMD)**: A method for measuring the distance between two text documents based on the Word2Vec embeddings. This lab demonstrates how to compute WMD to compare the similarity between sentences.
- **TF-IDF** (Term Frequency-Inverse Document Frequency): A statistical measure used to evaluate the importance of a word in a document relative to a corpus. The lab includes an example of calculating TF-IDF scores for a set of sentences.

#### Text Visualization:
- t-SNE (t-distributed Stochastic Neighbor Embedding): A technique for visualizing high-dimensional data by reducing it to two or three dimensions. This lab provides an example of using t-SNE to visualize word embeddings in a 2D space.

#### Practical Applications:
The lab concludes with a practical exercise where users are tasked with creating a simple chatbot. This chatbot will use word embeddings and similarity measures to respond to user queries based on a predefined set of questions and answers.

````python
tfIdfVectorizer=TfidfVectorizer(use_idf=True)
tfIdf = tfIdfVectorizer.fit_transform(dataset)
tfIdf[0].todense()
````

### t-SNE Visualization

````python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
````

````python
tokens = []
labels = []

for word in list(gwv.vocab.keys())[:1000]:
  tokens.append(gwv[word])
  labels.append(word)


tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
new_values = tsne_model.fit_transform(tokens)

x = []
y = []
for value in new_values:
    x.append(value[0])
    y.append(value[1])

plt.figure(figsize=(16, 16))
for i in range(len(x)):
    plt.scatter(x[i],y[i])
    plt.annotate(labels[i],
                  xy=(x[i], y[i]),
                  xytext=(5, 2),
                  textcoords='offset points',
                  ha='right',
                  va='bottom')
plt.show()
````

Example:

A collection of questions with answers:
```python
[
  {"q1" : "a1"},
  {"q2" : "a2"},
  {"q3" : "a3"},
  ...
  {"qn" : "an"},
]

```
Vectorize questions:
```python
[
  {"q1_vecotorized" : "a1"},
  {"q2_vecotorized" : "a2"},
  {"q3_vecotorized" : "a3"},
  ...
  {"qn_vecotorized" : "an"},
]
```
For the user question `q_user` we vectorize it and count the similarity with the questions in the database

```python
sim("q_user_vectorized","q1_vecotorized"), sim("q_user_vectorized","q2_vecotorized"), ... ("q_user_vectorized""qn_vecotorized")
```

We return the answer to the question from the base most similar to the user's question
$$
argmax(sim(q_{user}vectorized,q_{i}vecotorized))
$$

For calculating similarities, you can use e.g.:
https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.cosine.html but then we need to use argmin because
$$
cosine\ distance = 1 - cosine\ similarity
$$
